# Mi primer modelo: predecir ventas

## ¿Qué vamos a hacer?

Imagina que trabajas en una empresa y tu jefe te pregunta:

> *"Si invertimos $5,000 en marketing, ¿cuánto vamos a vender?"*

Para responder, vamos a usar los **datos históricos** de la empresa y trazar una **línea de tendencia** que nos permita hacer predicciones.

Eso es, en esencia, un **modelo de regresión lineal**: encontrar la mejor línea recta que describe la relación entre dos variables.

### En esta actividad aprenderás a:
1. Cargar y explorar datos reales de ventas
2. Visualizar la relación entre marketing y ventas
3. Construir tu primer modelo (una línea recta)
4. Hacer predicciones con ese modelo
5. Medir qué tan bien (o mal) predice

## 1. Preparación
Importamos las herramientas que necesitamos. Solo 3 librerías:

```
import pandas as pd        # Para trabajar con tablas de datos
import numpy as np         # Para cálculos matemáticos
import matplotlib.pyplot as plt  # Para hacer gráficos
```

## 2. Cargar los datos

Tenemos un archivo CSV con datos históricos de ventas. Cada fila representa un registro de venta con información como cuánto se invirtió en marketing, el precio del producto, la tienda, etc.

```
# Cargamos el archivo CSV en una tabla (DataFrame)
df = pd.read_csv("ventas_ml_clase2.csv")

# Veamos las primeras 5 filas
df.head()
```

```
# ¿Cuántos registros tenemos?
print(f"El dataset tiene {len(df)} registros (filas) y {df.shape[1]} columnas")
print()
print("Columnas disponibles:")
print("  - marketing:  cuánto se invirtió en publicidad ($)")
print("  - precio:     precio del producto ($)")
print("  - temporada:  trimestre del año (1, 2, 3 o 4)")
print("  - tienda:     ubicación de la tienda")
print("  - canal:      cómo se vendió (Tienda, Online, Mayorista)")
print("  - ventas:     lo que queremos predecir ($)")
```

## 3. Visualizar: ¿marketing y ventas están relacionados?

Antes de construir un modelo, lo primero es **mirar los datos**. Vamos a graficar la inversión en marketing contra las ventas para ver si hay una tendencia.

```
# Gráfico de dispersión (scatter plot)
# Cada punto es un registro: su posición horizontal es el marketing, la vertical son las ventas
plt.figure(figsize=(10, 6))
plt.scatter(df["marketing"], df["ventas"], alpha=0.4, color="steelblue", s=30)
plt.xlabel("Inversión en marketing ($)", fontsize=12)
plt.ylabel("Ventas ($)", fontsize=12)
plt.title("¿Más marketing = más ventas?", fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()
```

**¿Qué observas?** Los puntos tienden a subir hacia la derecha. Eso sugiere que a mayor inversión en marketing, las ventas tienden a ser mayores. No es perfecto (hay dispersión), pero hay una tendencia clara.

Nuestro objetivo es **trazar la mejor línea recta** que capture esa tendencia.

## 4. ¿Qué es una regresión lineal?

Una regresión lineal busca la **mejor línea recta** que pase "por el medio" de los puntos. La fórmula de cualquier línea recta es:

> **ventas = a + b × marketing**

Donde:
- **a** (intercepto): el valor de las ventas cuando marketing es $0. Es el "punto de partida" de la línea.
- **b** (pendiente): cuánto suben las ventas por cada $1 extra de marketing. Es la "inclinación" de la línea.

El modelo lo que hace es encontrar los valores de **a** y **b** que producen la línea que mejor se ajusta a los datos.

## 5. Construir el modelo (encontrar la mejor línea)

Usamos `np.polyfit()` que encuentra los valores de **a** y **b** que minimizan la distancia entre la línea y los puntos.

```
# Extraemos las dos columnas que nos interesan
x = df["marketing"].values   # Variable predictora (lo que usamos para predecir)
y = df["ventas"].values      # Variable objetivo (lo que queremos predecir)

# np.polyfit encuentra la mejor línea recta
# El "1" significa que queremos un polinomio de grado 1 (= una línea recta)
b, a = np.polyfit(x, y, 1)

print("¡Modelo listo!")
print()
print(f"La fórmula de nuestro modelo es:")
print(f"  ventas = {a:.2f} + {b:.4f} × marketing")
print()
print(f"¿Qué significa?")
print(f"  - Punto de partida (a): cuando marketing = $0, las ventas base son ${a:.2f}")
print(f"  - Pendiente (b): por cada $1 extra en marketing, las ventas suben ${b:.4f}")
print(f"  - O dicho de otro modo: por cada $1,000 extra en marketing, las ventas suben ${b*1000:.2f}")
```

## 6. Dibujar la línea del modelo sobre los datos

Ahora vamos a superponer nuestra línea recta sobre los puntos para ver qué tan bien se ajusta.

```
# Creamos puntos sobre la línea para dibujarla
# Tomamos desde el mínimo hasta el máximo de marketing
x_linea = np.array([x.min(), x.max()])
y_linea = a + b * x_linea    # Aplicamos la fórmula: ventas = a + b × marketing

# Graficamos los datos + la línea del modelo
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.4, color="steelblue", s=30, label="Datos reales")
plt.plot(x_linea, y_linea, color="red", linewidth=2, label="Modelo (línea de predicción)")
plt.xlabel("Inversión en marketing ($)", fontsize=12)
plt.ylabel("Ventas ($)", fontsize=12)
plt.title("Nuestro modelo: la mejor línea recta", fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.show()
```

La línea roja es nuestro modelo. Pasa "por el medio" de los puntos y captura la tendencia general: más marketing → más ventas.

## 7. Hacer predicciones

Ahora podemos usar la fórmula para responder preguntas del negocio. Solo sustituimos el valor de marketing en la fórmula.

```
# Definimos algunos escenarios de inversión en marketing
escenarios = [1000, 3000, 5000, 7000, 10000]

print("=" * 50)
print("   PREDICCIONES DE VENTAS")
print("=" * 50)
print(f"   Fórmula: ventas = {a:.2f} + {b:.4f} × marketing")
print("-" * 50)

for mkt in escenarios:
    # La predicción es simplemente sustituir en la fórmula
    prediccion = a + b * mkt
    print(f"   Si marketing = ${mkt:>6,}  →  ventas predichas = ${prediccion:,.2f}")

print("=" * 50)
```

**Así de simple es usar el modelo.** Solo sustituimos el valor de marketing en la fórmula y obtenemos una predicción de ventas.

## 8. ¿Qué tan bien predice? Midamos el error

Un modelo nunca es perfecto. Siempre habrá diferencias entre lo que predice y lo que realmente pasó. A esas diferencias les llamamos **errores**.

La pregunta clave es: **¿qué tan grandes son esos errores?**

```
# Calculamos la predicción para TODOS los registros
predicciones = a + b * x

# El error de cada registro = valor real - valor predicho
errores = y - predicciones

# Veamos algunos ejemplos
print("Ejemplos de predicciones vs realidad:")
print("-" * 65)
# :>12 significa "alinear a la derecha en un espacio de 12 caracteres"
print(f"  {'Marketing':>12}  {'Venta real':>12}  {'Predicción':>12}  {'Error':>10}")
print("-" * 65)
for i in range(8):
    print(f"  ${x[i]:>10,.2f}  ${y[i]:>10,.2f}  ${predicciones[i]:>10,.2f}  ${errores[i]:>+9,.2f}")
print("-" * 65)
print()
print("Error positivo (+) = el modelo predijo MENOS de lo real (se quedó corto)")
print("Error negativo (-) = el modelo predijo MÁS de lo real (se pasó)")
```

```
# --- Error Absoluto Medio (MAE) ---
# Tomamos el valor absoluto de cada error (ignoramos el signo +/-)
# y calculamos el promedio
mae = np.mean(np.abs(errores))

# Promedio de ventas reales (para dar contexto)
promedio_ventas = y.mean()

print("=" * 50)
print("   ¿QUÉ TAN BIEN PREDICE EL MODELO?")
print("=" * 50)
print(f"   Error promedio (MAE): ${mae:.2f}")
print(f"   Ventas promedio:      ${promedio_ventas:.2f}")
print(f"   Error relativo:       {mae/promedio_ventas*100:.1f}%")
print("=" * 50)
print()
print(f"En palabras simples:")
print(f"  Cada vez que el modelo predice ventas, se equivoca")
print(f"  en promedio por ${mae:.2f} (un {mae/promedio_ventas*100:.1f}% de la venta típica).")
```

## 9. Visualizar los errores

Cada línea vertical roja conecta un punto real con lo que el modelo predijo. Líneas cortas = buena predicción. Líneas largas = el modelo se equivocó bastante.

```
# Tomamos una muestra de 50 puntos para que el gráfico sea legible
np.random.seed(42)
muestra = np.random.choice(len(x), size=50, replace=False)
muestra.sort()

plt.figure(figsize=(10, 6))

# Línea del modelo
plt.plot(x_linea, y_linea, color="red", linewidth=2, label="Modelo")

# Puntos reales (muestra)
plt.scatter(x[muestra], y[muestra], color="steelblue", s=40, zorder=3, label="Datos reales")

# Líneas de error: conectan cada punto real con la línea del modelo
for i in muestra:
    plt.plot([x[i], x[i]], [y[i], predicciones[i]], color="red", alpha=0.4, linewidth=1)

plt.xlabel("Inversión en marketing ($)", fontsize=12)
plt.ylabel("Ventas ($)", fontsize=12)
plt.title("Errores del modelo (líneas rojas = diferencia entre predicción y realidad)", fontsize=13)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.show()
```

**¿Qué observas?** Hay bastante dispersión: muchos puntos están lejos de la línea. Esto nos dice que el marketing **no es el único factor** que afecta las ventas. Hay otros factores (precio, temporada, tienda, canal) que también influyen y que nuestro modelo simple no está considerando.

Un modelo con más variables podría ser más preciso, pero este modelo con una sola variable ya nos da una idea útil de la tendencia general.

## 10. Simulación: ¿qué pasa si aumentamos el marketing?

Una de las ventajas de tener un modelo es poder hacer simulaciones "¿qué pasaría si...?"

```
# Escenario: inversión actual vs +20%
inversion_actual = 5000
inversion_nueva = inversion_actual * 1.20   # +20%

venta_actual = a + b * inversion_actual
venta_nueva = a + b * inversion_nueva
diferencia = venta_nueva - venta_actual

print("=" * 55)
print("   SIMULACIÓN: ¿Qué pasa si subimos marketing un 20%?")
print("=" * 55)
print(f"   Inversión actual:     ${inversion_actual:>10,}")
print(f"   Inversión nueva:      ${inversion_nueva:>10,.0f}  (+20%)")
print("-" * 55)
print(f"   Ventas predichas (antes):  ${venta_actual:>10,.2f}")
print(f"   Ventas predichas (después): ${venta_nueva:>10,.2f}")
print(f"   Incremento en ventas:       ${diferencia:>10,.2f}")
print("=" * 55)
print()
print(f"Invertir ${inversion_nueva - inversion_actual:,.0f} más en marketing")
print(f"generaría un incremento estimado de ${diferencia:,.2f} en ventas.")
```

## 11. Resumen: ¿qué aprendimos?

| Concepto | Qué significa |
|----------|---------------|
| **Regresión lineal** | Encontrar la mejor línea recta que describe la relación entre dos variables |
| **Fórmula** | `ventas = a + b × marketing` — una ecuación simple para predecir |
| **Intercepto (a)** | El valor base de ventas cuando marketing es $0 |
| **Pendiente (b)** | Cuánto cambian las ventas por cada $1 extra en marketing |
| **Error (MAE)** | En promedio, cuánto se equivoca el modelo en cada predicción |

### Limitaciones de este modelo
- Solo usamos **una variable** (marketing). En la realidad, las ventas dependen de muchos factores.
- Un modelo con más variables (precio, temporada, tienda, canal) sería más preciso.
- La regresión lineal asume una relación en línea recta, pero no siempre es así.
